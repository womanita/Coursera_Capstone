<font size=6 font face='garamond' font color='blue'> And now, I remove the data where 'Borough' was not assigned...</font>

```python
TorontoDataFrame=TorontoDataFrame[TorontoDataFrame['Borough']!='Not assigned']
TorontoDataFrame
```

<font face='times new roman' font size=4> And now I replace the missing Neighbourhood names with their Borough names.  This was pretty simple, because I found only one example in the whole dataframe: in Queen's Park.

```python
TorontoDataFrame=TorontoDataFrame.replace({'Not assigned':"Queen's Park"})
TorontoDataFrame
```

<font face='times new roman' font size=4> And now, I group all the Neighbourhood names together by Postcode and Borough.

```python
TorontoDataFrame=TorontoDataFrame.groupby(['Postcode',"Borough"]).agg(lambda x: ", ".join(x))
TorontoDataFrame
```

<font face='times new roman' font size=4> The aggregate function worked wonderfully for this purpose, but left one odd problem: it changed the Postcode and Borough columns into the index of the dataframe.  This can be corrected by using the reset_index function.

```python
TorontoDataFrame=TorontoDataFrame.reset_index(level=['Postcode','Borough'])
TorontoDataFrame
```

<font face='Times New Roman' font color='green' font size=8> And now on to Question 2... </font>

<font face='times new roman' font size=4> First, I download the columns of latitude and longitude as a csv.

```python
latlongdf=pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
latlongdf=latlongdf[['Latitude','Longitude']]
```

<font face='times new roman' font size=4> And then, I use the concat function to join the new columns to my dataframe.

```python
joinedDF=pd.concat([TorontoDataFrame,latlongdf],axis=1,sort=False)
joinedDF
```

<font face='garamond' font color='green' font size=6> And now to Question 3 </font>

<font face='times new roman' font size=4> First, I created a Folium map of all my data points.

```python
map_toronto = folium.Map(location=[43.6532, -79.3832], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(joinedDF['Latitude'], joinedDF['Longitude'], joinedDF['Borough'], joinedDF['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

<font face='times new roman' font size=4> And now, I do some analysis with the data grouped by different neighbourhoods.  What I will do is create a map with all the Boroughs with 'Toronto' in their names, each one in a different color.  Following are the dataframes for each group...

```python
NamedToronto=joinedDF.groupby('Borough')
NamedToronto
```

```python
NamedDowntown=NamedToronto.get_group('Downtown Toronto')
NamedDowntown
```

```python
NamedCentral=NamedToronto.get_group('Central Toronto')
NamedCentral
```

```python
NamedEast=NamedToronto.get_group('East Toronto')
NamedEast
```

```python
NamedWest=NamedToronto.get_group('West Toronto')
NamedWest
```

```python
map_toronto = folium.Map(location=[43.6532, -79.3832], zoom_start=10)

# add markers to my new map, grouped by boroughs with 'Toronto' in their names.
for lat, lng, borough, neighborhood in zip(NamedDowntown['Latitude'], NamedDowntown['Longitude'], NamedDowntown['Borough'], NamedDowntown['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='#ffc7d2',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
    
for lat, lng, borough, neighborhood in zip(NamedCentral['Latitude'], NamedCentral['Longitude'], NamedCentral['Borough'], NamedCentral['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='yellow',
        fill=True,
        fill_color='#ffff92',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto) 

for lat, lng, borough, neighborhood in zip(NamedEast['Latitude'], NamedEast['Longitude'], NamedEast['Borough'], NamedEast['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='green',
        fill=True,
        fill_color='#6df758',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto) 

for lat, lng, borough, neighborhood in zip(NamedWest['Latitude'], NamedWest['Longitude'], NamedWest['Borough'], NamedWest['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#58f7f2',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto) 

map_toronto
```

<font face='times new roman' font size=4>Wow, I did it!  As you can see, there is a very nice and visible division between these boroughs.  An interesting thing to note is that<font color='orange'> Central Toronto</font>, in yellow, is north of <font color='red'>Downtown Toronto</font>, in red.  This could mean that the historic Downtown formed along the waterfront before the growth of the city's central business district, which grew in the north.  Another interesting thing is how <font color='green'>East</font> and <font color='blue'>West</font> Toronto don't extend north of <font color='red'>Downtown</font>, and <font color='orange'>Central</font> Toronto's boundaries do not extend any further east or west of <font color='red'>Downtown's</font> boundaries.  This makes it look like the Boroughs with 'Toronto' in their names were named in relation to Downtown.</font>

<font face='times new roman' font size=4>A very quick web search for the repeated names(such as Davisville and Forest Hill in Central Toronto, Parkdale in West Toronto and Beaches in East Toronto) in each Borough revealed that Central Toronto seems to have more wealthy neighborhoods, while West Toronto and East Toronto have more lower-income neighborhoods.  This seems to support my conclusion that Central Toronto grew as a business area after the old downtown.
